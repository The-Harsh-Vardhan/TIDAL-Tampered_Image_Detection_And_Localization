```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# ============================================================
# CONFIGURATION
# ============================================================
DATASET_PATH = '/kaggle/input/datasets/manjilkarki/deepfake-and-real-images/Dataset'
IMG_SIZE     = 224          # ← Reduced from 299 (saves ~40% compute)
BATCH_SIZE   = 32
EPOCHS       = 30
SEED         = 42

TRAIN_DIR = os.path.join(DATASET_PATH, 'Train')
VAL_DIR   = os.path.join(DATASET_PATH, 'Validation')
TEST_DIR  = os.path.join(DATASET_PATH, 'Test')

print("TensorFlow Version:", tf.__version__)
print("GPU Available     :", tf.config.list_physical_devices('GPU'))

# ============================================================
# DATA GENERATORS
# ============================================================
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.15,
    horizontal_flip=True,
    fill_mode='nearest'
)
val_test_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(
    TRAIN_DIR, target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE, class_mode='binary', seed=SEED
)
val_gen = val_test_datagen.flow_from_directory(
    VAL_DIR, target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE, class_mode='binary', seed=SEED
)
test_gen = val_test_datagen.flow_from_directory(
    TEST_DIR, target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE, class_mode='binary',
    shuffle=False, seed=SEED
)

print(f"\nClass Indices : {train_gen.class_indices}")
print(f"Train Samples : {train_gen.n}")
print(f"Val   Samples : {val_gen.n}")
print(f"Test  Samples : {test_gen.n}")


# ============================================================
# BUILDING BLOCKS  (filter counts ~50% of original)
# ============================================================

# ── V1 Module: Classic parallel branches (no BN — original style) ──
def inception_v1_module(x, f1, f3r, f3, f5r, f5, fpool, name):
    b1 = layers.Conv2D(f1,   (1,1), padding='same', activation='relu', name=f'{name}_b1')(x)

    b2 = layers.Conv2D(f3r,  (1,1), padding='same', activation='relu', name=f'{name}_b2r')(x)
    b2 = layers.Conv2D(f3,   (3,3), padding='same', activation='relu', name=f'{name}_b2')(b2)

    b3 = layers.Conv2D(f5r,  (1,1), padding='same', activation='relu', name=f'{name}_b3r')(x)
    b3 = layers.Conv2D(f5,   (5,5), padding='same', activation='relu', name=f'{name}_b3')(b3)

    b4 = layers.MaxPooling2D((3,3), strides=1, padding='same', name=f'{name}_b4p')(x)
    b4 = layers.Conv2D(fpool,(1,1), padding='same', activation='relu', name=f'{name}_b4')(b4)

    return layers.Concatenate(name=f'{name}_concat')([b1, b2, b3, b4])


# ── V2 Module: 5×5 → two 3×3 + BatchNorm ──
def inception_v2_module(x, f1, f3r, f3, fdr, fd, fpool, name):
    b1 = layers.Conv2D(f1,  (1,1), padding='same', activation='relu', name=f'{name}_b1')(x)
    b1 = layers.BatchNormalization(name=f'{name}_b1_bn')(b1)

    b2 = layers.Conv2D(f3r, (1,1), padding='same', activation='relu', name=f'{name}_b2r')(x)
    b2 = layers.BatchNormalization(name=f'{name}_b2r_bn')(b2)
    b2 = layers.Conv2D(f3,  (3,3), padding='same', activation='relu', name=f'{name}_b2')(b2)
    b2 = layers.BatchNormalization(name=f'{name}_b2_bn')(b2)

    b3 = layers.Conv2D(fdr, (1,1), padding='same', activation='relu', name=f'{name}_b3r')(x)
    b3 = layers.BatchNormalization(name=f'{name}_b3r_bn')(b3)
    b3 = layers.Conv2D(fd,  (3,3), padding='same', activation='relu', name=f'{name}_b3a')(b3)
    b3 = layers.BatchNormalization(name=f'{name}_b3a_bn')(b3)
    b3 = layers.Conv2D(fd,  (3,3), padding='same', activation='relu', name=f'{name}_b3b')(b3)
    b3 = layers.BatchNormalization(name=f'{name}_b3b_bn')(b3)

    b4 = layers.AveragePooling2D((3,3), strides=1, padding='same', name=f'{name}_b4p')(x)
    b4 = layers.Conv2D(fpool,(1,1), padding='same', activation='relu', name=f'{name}_b4')(b4)
    b4 = layers.BatchNormalization(name=f'{name}_b4_bn')(b4)

    return layers.Concatenate(name=f'{name}_concat')([b1, b2, b3, b4])


# ── V3 Module: Asymmetric 1×n + n×1 factorization ──
def inception_v3_module(x, f1, f3r, f3, fdr, fd, fpool, n=7, name='v3mod'):
    b1 = layers.Conv2D(f1,  (1,1), padding='same', activation='relu', name=f'{name}_b1')(x)
    b1 = layers.BatchNormalization(name=f'{name}_b1_bn')(b1)

    b2 = layers.Conv2D(f3r, (1,1), padding='same', activation='relu', name=f'{name}_b2r')(x)
    b2 = layers.BatchNormalization(name=f'{name}_b2r_bn')(b2)
    b2 = layers.Conv2D(f3,  (1,n), padding='same', activation='relu', name=f'{name}_b2a')(b2)
    b2 = layers.Conv2D(f3,  (n,1), padding='same', activation='relu', name=f'{name}_b2b')(b2)
    b2 = layers.BatchNormalization(name=f'{name}_b2_bn')(b2)

    b3 = layers.Conv2D(fdr, (1,1), padding='same', activation='relu', name=f'{name}_b3r')(x)
    b3 = layers.BatchNormalization(name=f'{name}_b3r_bn')(b3)
    b3 = layers.Conv2D(fd,  (1,n), padding='same', activation='relu', name=f'{name}_b3a')(b3)
    b3 = layers.Conv2D(fd,  (n,1), padding='same', activation='relu', name=f'{name}_b3b')(b3)
    b3 = layers.Conv2D(fd,  (1,n), padding='same', activation='relu', name=f'{name}_b3c')(b3)
    b3 = layers.Conv2D(fd,  (n,1), padding='same', activation='relu', name=f'{name}_b3d')(b3)
    b3 = layers.BatchNormalization(name=f'{name}_b3_bn')(b3)

    b4 = layers.AveragePooling2D((3,3), strides=1, padding='same', name=f'{name}_b4p')(x)
    b4 = layers.Conv2D(fpool,(1,1), padding='same', activation='relu', name=f'{name}_b4')(b4)
    b4 = layers.BatchNormalization(name=f'{name}_b4_bn')(b4)

    return layers.Concatenate(name=f'{name}_concat')([b1, b2, b3, b4])



def build_unified_inception(input_shape=(224, 224, 3)):
    inputs = keras.Input(shape=input_shape, name='input')

    # ── STEM (slimmed: 32→32→96 instead of 32→64→192) ─────────
    x = layers.Conv2D(32, (3,3), strides=2, padding='same', activation='relu', name='stem_c1')(inputs)
    x = layers.BatchNormalization(name='stem_bn1')(x)
    x = layers.Conv2D(32, (3,3), padding='same', activation='relu', name='stem_c2')(x)
    x = layers.BatchNormalization(name='stem_bn2')(x)
    x = layers.MaxPooling2D((3,3), strides=2, padding='same', name='stem_pool')(x)
    x = layers.Conv2D(96, (3,3), padding='same', activation='relu', name='stem_c3')(x)
    x = layers.BatchNormalization(name='stem_bn3')(x)
    x = layers.MaxPooling2D((3,3), strides=2, padding='same', name='stem_pool2')(x)

    # ── STAGE 1: INCEPTION V1  (2 blocks, ~half filters) ───────
    #   Original: (64,96,128,16,32,32) → (128,128,192,32,96,64)
    #   Reduced:  (32,48, 64, 8,16,16) → ( 64, 64, 96,16,48,32)
    x = inception_v1_module(x,  32, 48,  64,  8, 16, 16, name='v1_m1')
    x = inception_v1_module(x,  64, 64,  96, 16, 48, 32, name='v1_m2')
    x = layers.MaxPooling2D((3,3), strides=2, padding='same', name='v1_pool')(x)

    skip_v1 = layers.GlobalAveragePooling2D(name='skip_v1_gap')(x)
    skip_v1 = layers.Dense(64, activation='relu', name='skip_v1_fc')(skip_v1)

    # ── STAGE 2: INCEPTION V2  (2 blocks instead of 3) ─────────
    #   Original: (192,96,208,16,48,64) × 3
    #   Reduced:  ( 96,48,104, 8,24,32) × 2
    x = inception_v2_module(x,  96, 48, 104,  8, 24, 32, name='v2_m1')
    x = inception_v2_module(x,  80, 56, 112, 12, 32, 32, name='v2_m2')
    x = layers.MaxPooling2D((3,3), strides=2, padding='same', name='v2_pool')(x)

    skip_v2 = layers.GlobalAveragePooling2D(name='skip_v2_gap')(x)
    skip_v2 = layers.Dense(64, activation='relu', name='skip_v2_fc')(skip_v2)

    # ── STAGE 3: INCEPTION V3  (3 blocks instead of 4) ─────────
    #   Original: 4 blocks with large filters (192–320)
    #   Reduced:  3 blocks with smaller filters (96–160)
    x = inception_v3_module(x,  96, 48, 104,  8, 24, 32, n=3, name='v3_m1')
    x = inception_v3_module(x,  80, 56, 112, 12, 32, 32, n=3, name='v3_m2')
    x = inception_v3_module(x, 128, 80, 160, 16, 64, 64, n=5, name='v3_m3')

    # ── FUSION HEAD (128 → 64 → 1 instead of 256 → 128 → 1) ───
    x = layers.GlobalAveragePooling2D(name='final_gap')(x)
    x = layers.Dense(128, activation='relu', name='final_fc')(x)
    x = layers.BatchNormalization(name='final_bn')(x)

    # Multi-scale concat: V3(128) + skip_v1(64) + skip_v2(64) = 256 total
    x = layers.Concatenate(name='multiscale_fusion')([x, skip_v1, skip_v2])

    x = layers.Dense(128, activation='relu', name='fusion_fc1')(x)
    x = layers.Dropout(0.5, name='dropout1')(x)
    x = layers.Dense(64,  activation='relu', name='fusion_fc2')(x)
    x = layers.Dropout(0.3, name='dropout2')(x)
    outputs = layers.Dense(1, activation='sigmoid', name='output')(x)

    return Model(inputs, outputs, name='Unified_InceptionV1_V2_V3_Slim')


# ============================================================
# BUILD & COMPILE
# ============================================================
model = build_unified_inception()
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss='binary_crossentropy',
    metrics=[
        'accuracy',
        tf.keras.metrics.AUC(name='auc'),
        tf.keras.metrics.Precision(name='precision'),
        tf.keras.metrics.Recall(name='recall')
    ]
)

model.summary()
print(f"\nTotal Parameters: {model.count_params():,}")

# ============================================================
# CALLBACKS
# ============================================================
callbacks = [
    ModelCheckpoint(
        'unified_inception_slim_best.keras',
        monitor='val_accuracy', save_best_only=True, verbose=1
    ),
    EarlyStopping(
        monitor='val_accuracy', patience=8,
        restore_best_weights=True, verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss', factor=0.3,
        patience=3, min_lr=1e-7, verbose=1
    )
]

# ============================================================
# TRAIN
# ============================================================
print("\n" + "="*60)
print("  TRAINING SLIM UNIFIED INCEPTION V1 + V2 + V3 MODEL")
print("="*60)

history = model.fit(
    train_gen,
    epochs=EPOCHS,
    validation_data=val_gen,
    callbacks=callbacks,
    verbose=1
)

# ============================================================
# PLOTS
# ============================================================
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Slim Unified Inception V1+V2+V3 — Training History',
             fontsize=14, fontweight='bold')

metrics_to_plot = [
    ('accuracy', 'Accuracy'),
    ('loss',     'Loss'),
    ('auc',      'AUC'),
    ('precision','Precision')
]
for ax, (m, title) in zip(axes.flatten(), metrics_to_plot):
    ax.plot(history.history[m],           label=f'Train {title}', color='royalblue', lw=2)
    ax.plot(history.history[f'val_{m}'],  label=f'Val {title}',   color='tomato',    lw=2)
    ax.set_title(title, fontsize=12)
    ax.legend(); ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('training_history.png', dpi=150)
plt.show()

# ============================================================
# EVALUATION
# ============================================================
print("\n" + "="*60)
print("  FINAL EVALUATION ON TEST SET")
print("="*60)

results = model.evaluate(test_gen, verbose=1)
print(f"\n  Test Loss      : {results[0]:.4f}")
print(f"  Test Accuracy  : {results[1]:.4f}  ({results[1]*100:.2f}%)")
print(f"  Test AUC       : {results[2]:.4f}")
print(f"  Test Precision : {results[3]:.4f}")
print(f"  Test Recall    : {results[4]:.4f}")

preds_prob = model.predict(test_gen, verbose=1).flatten()
preds      = (preds_prob > 0.5).astype(int)
labels     = test_gen.classes

print("\n  Classification Report:")
print(classification_report(labels, preds, target_names=['Fake', 'Real']))

# Confusion Matrix
cm = confusion_matrix(labels, preds)
plt.figure(figsize=(7, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Fake', 'Real'],
            yticklabels=['Fake', 'Real'],
            linewidths=1, linecolor='black')
plt.title('Confusion Matrix — Slim Unified Inception V1+V2+V3',
          fontsize=13, fontweight='bold')
plt.ylabel('Actual', fontsize=12)
plt.xlabel('Predicted', fontsize=12)
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=150)
plt.show()


model.save('unified_inception_slim.keras')
print("Model saved: unified_inception_slim.keras")
```
